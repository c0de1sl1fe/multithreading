```python
import matplotlib.pyplot as plt

def readStats(path: str):
    x = []
    y = []
    with open(path, "r") as f:
        line = f.readline()
        while line: 
            tmp =[int(x) for x in line.split(" ")]
            print(tmp)
            x.append(tmp[0])
            y.append(tmp[1])
            line = f.readline()
    return (x,y)

x = []
y = []
x, y = readStats("D:\git\multithreading\stats_lab1.txt")
x.sort()
y.sort()
plt.plot(x, y, label='lab1')


x_2 = []
y_2 = []
x_2, y_2 = readStats("D:\git\multithreading\stats_lab2.txt")
x_2.sort()
y_2.sort()
plt.plot(x_2, y_2, label='lab2')




plt.title("Time dependence on size")
plt.xlabel("Size")
plt.ylabel("Time")
plt.legend()
plt.show()
```
