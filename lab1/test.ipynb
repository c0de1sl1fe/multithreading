```python
import matplotlib.pyplot as plt

def readStats(path: str):
    x = []
    y = []
    with open(path, "r") as f:
        line = f.readline()
        while line: 
            tmp =[int(x) for x in line.split(" ")]
            print(tmp)
            x.append(tmp[0])
            y.append(tmp[1])
            line = f.readline()
    return (x,y)
x = []
y = []
x, y = readStats("D:\Git\multithreading\stats.txt")
x.sort()
y.sort()



plt.plot(x, y, color="blue")
plt.show()
```
